# Problem: Find the Equation of a Line

The line passes through the points $A(1, 2)$ and $B(3, 4)$. Find the equation of the line.

---

## Solution

### 1. Calculate the Slope
The slope $m$ of a line passing through two points $(x_1, y_1)$ and $(x_2, y_2)$ is given by:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substituting the given points $A(1, 2)$ and $B(3, 4)$:

$$
m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
$$

### 2. Find the y-intercept ($b$)
The equation of a line is:

$$
y = mx + b
$$

Here, $m = 1$. To find $b$, substitute one of the points (e.g., $A(1, 2)$) into the equation:

$$
2 = 1(1) + b
$$

$$
b = 1
$$

### 3. Write the Equation of the Line
Substitute $m = 1$ and $b = 1$ into the line equation:

$$
y = x + 1
$$

---

# Problem: Find the Equation of the Line

The line passes through point \( A(1, 2) \) and is parallel to the line \( y = 2x + 3 \). Find the equation of the line.

---

## Solution

### 1. Identify the Slope of the Given Line

The given line is \( y = 2x + 3 \). In this equation, the slope \( m \) is the coefficient of \( x \), so:

$$
m = 2
$$

Since the line we need to find is parallel to this line, it will have the same slope.

### 2. Use the Point-Slope Formula

The point-slope formula for the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

Where \( m \) is the slope, and \( (x_1, y_1) \) is the point the line passes through. We know the line passes through point \( A(1, 2) \), so \( x_1 = 1 \) and \( y_1 = 2 \), and \( m = 2 \).

Substituting these values into the point-slope formula:

$$
y - 2 = 2(x - 1)
$$

### 3. Simplify the Equation

Now, simplify the equation to get the line equation in slope-intercept form:

$$
y - 2 = 2x - 2
$$

Add 2 to both sides:

$$
y = 2x
$$

---

### Final Answer

The equation of the line is:

$$
y = 2x
$$


````python
import matplotlib.pyplot as plt
import numpy as np

# Verilen doru: y = 2x + 3
# Paralel doru: y = 2x (A(1, 2) noktasndan geecek)

# x deerleri iin bir aralk tanmla
x = np.linspace(-10, 10, 100)

# Verilen dorunun denklemi
y1 = 2 * x + 3

# Paralel dorunun denklemi (A noktasndan geen)
y2 = 2 * x + (2 - 2 * 1)  # y - y1 = m(x - x1) kullanarak kesiim noktas hesapland

# Grafii iz
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="y = 2x + 3 (Given Line)", color="blue")
plt.plot(x, y2, label="y = 2x (Parallel Line)", color="orange")

# Noktay iaretle
plt.scatter(1, 2, color="red", label="A(1, 2)")

# Grafii zelletir
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Parallel Line Passing Through Point A(1, 2)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

# Problem: Find the Equation of the Line

The line passes through point \( A(1, 2) \) and is perpendicular to the line \( y = 2x + 3 \). Find the equation of the line.

---

## Solution

### 1. Identify the Slope of the Given Line

The given line is \( y = 2x + 3 \). In this equation, the slope \( m \) is the coefficient of \( x \), so:

$$
m_1 = 2
$$

### 2. Find the Slope of the Perpendicular Line

The slope of a line perpendicular to another is the negative reciprocal of its slope. Therefore, the slope of the required line is:

$$
m_2 = -\frac{1}{m_1} = -\frac{1}{2}
$$

### 3. Use the Point-Slope Formula

The point-slope formula for the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

Where \( m \) is the slope, and \( (x_1, y_1) \) is the point the line passes through. We know the line passes through point \( A(1, 2) \), so \( x_1 = 1 \) and \( y_1 = 2 \), and \( m = -\frac{1}{2} \).

Substituting these values into the point-slope formula:

$$
y - 2 = -\frac{1}{2}(x - 1)
$$

### 4. Simplify the Equation

Now, simplify the equation to get the line equation in slope-intercept form:

$$
y - 2 = -\frac{1}{2}x + \frac{1}{2}
$$

Add 2 to both sides:

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

---

### Final Answer

The equation of the line is:

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$


````python
import matplotlib.pyplot as plt
import numpy as np

# Verilen doru: y = 2x + 3
# Dik doru: Eimi -1/2 olan ve A(1, 2) noktasndan geen doru

# x deerleri iin bir aralk tanmla
x = np.linspace(-10, 10, 100)

# Verilen dorunun denklemi
y1 = 2 * x + 3

# Dik dorunun denklemi: y = -1/2 * x + b
# A(1, 2) noktasndan getii iin b'yi hesaplyoruz:
# 2 = -1/2 * 1 + b => b = 2 + 1/2 = 5/2
y2 = -0.5 * x + 2.5

# Grafii iz
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="y = 2x + 3 (Given Line)", color="blue")
plt.plot(x, y2, label="y = -1/2 * x + 5/2 (Perpendicular Line)", color="orange")

# Noktay iaretle
plt.scatter(1, 2, color="red", label="A(1, 2)")

# Grafii zelletir
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Perpendicular Line Passing Through Point A(1, 2)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
````

# Problem: Find the Intersection Point and the Angle Between Two Lines

We are given two lines:
- \( y = 2x + 3 \)
- \( y = 3x + 2 \)

Find the intersection point of these lines and calculate the angle between them.

---

## Solution

### 1. Find the Intersection Point

To find the intersection point, we solve the system of equations:

$$
y = 2x + 3
$$
$$
y = 3x + 2
$$

Set the two equations equal to each other:

$$
2x + 3 = 3x + 2
$$

Solve for \( x \):

$$
2x - 3x = 2 - 3
$$
$$
-x = -1
$$
$$
x = 1
$$

Now substitute \( x = 1 \) into either of the original equations to find \( y \). Using \( y = 2x + 3 \):

$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is \( (1, 5) \).

### 2. Calculate the Angle Between the Two Lines

The formula to calculate the angle \( \theta \) between two lines is:

$$
\tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1m_2} \right|
$$

Where \( m_1 \) and \( m_2 \) are the slopes of the two lines.

For the line \( y = 2x + 3 \), the slope is \( m_1 = 2 \), and for the line \( y = 3x + 2 \), the slope is \( m_2 = 3 \).

Substitute these values into the formula:

$$
\tan(\theta) = \left| \frac{2 - 3}{1 + (2)(3)} \right| = \left| \frac{-1}{1 + 6} \right| = \left| \frac{-1}{7} \right| = \frac{1}{7}
$$

Now, calculate the angle:

$$
\theta = \tan^{-1}\left(\frac{1}{7}\right)
$$

Using a calculator:

$$
\theta \approx 8.13^\circ
$$

---

### Final Answer

- The intersection point of the lines is \( (1, 5) \).
- The angle between the two lines is approximately \( 8.13^\circ \).


````python
import numpy as np
import matplotlib.pyplot as plt

# Define the slopes of the lines
m1 = 2  # Slope of the first line (y = 2x + 3)
m2 = 3  # Slope of the second line (y = 3x + 2)

# 1. Find the intersection point

# Set the two equations equal to each other:
# 2x + 3 = 3x + 2 -> 2x - 3x = 2 - 3 -> -x = -1 -> x = 1
x_intersection = 1

# Substitute x = 1 into one of the equations to find y
y_intersection = 2 * x_intersection + 3  # Using y = 2x + 3

# Intersection point
intersection_point = (x_intersection, y_intersection)

# 2. Calculate the angle between the lines
# tan(theta) = |(m1 - m2) / (1 + m1*m2)|
tan_theta = abs((m1 - m2) / (1 + m1 * m2))

# Calculate the angle in degrees
theta = np.arctan(tan_theta) * (180 / np.pi)

# 3. Plot the lines and the intersection point
x = np.linspace(-10, 10, 100)

# Line equations
y1 = m1 * x + 3  # y = 2x + 3
y2 = m2 * x + 2  # y = 3x + 2

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="y = 2x + 3", color="blue")
plt.plot(x, y2, label="y = 3x + 2", color="green")

# Plot the intersection point
plt.scatter(x_intersection, y_intersection, color="red", zorder=5)
plt.text(x_intersection + 0.2, y_intersection, f"({x_intersection}, {y_intersection})", fontsize=12, color="red")

# Customize the plot
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title(f"Intersection Point: ({x_intersection}, {y_intersection})\nAngle between lines: {theta:.2f}")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(-10, 10)
plt.ylim(-10, 20)
plt.show()

# Final output
print(f"Intersection Point: ({x_intersection}, {y_intersection})")
print(f"Angle between the lines: {theta:.2f} degrees")
````

## Problem: Equation of the Line

**Given**:  
A line passes through point \( A(1, 2) \) and is parallel to the vector \( \vec{v} = [2, 3] \).

**Find**: The equation of the line.

---

## Solution

### 1. **Direction of the Line**

The direction of the line is given by the vector \( \vec{v} = [2, 3] \). This means that the slope of the line is the ratio of the \( y \)-component to the \( x \)-component of the vector, which is:

$$
m = \frac{3}{2}
$$

### 2. **Point-Slope Formula**

The point-slope formula for the equation of a line is:

$$
y - y_1 = m(x - x_1)
$$

Where:
- \( m \) is the slope,
- \( (x_1, y_1) \) is the point the line passes through.

We know the line passes through point \( A(1, 2) \), so \( x_1 = 1 \) and \( y_1 = 2 \), and \( m = \frac{3}{2} \).

### 3. **Substitute Values**

Substitute the known values into the point-slope formula:

$$
y - 2 = \frac{3}{2}(x - 1)
$$

### 4. **Simplify the Equation**

Now, simplify the equation:

$$
y - 2 = \frac{3}{2}x - \frac{3}{2}
$$

Add 2 to both sides to get the equation in slope-intercept form:

$$
y = \frac{3}{2}x + \frac{1}{2}
$$

---

### Final Answer

The equation of the line is:

$$
y = \frac{3}{2}x + \frac{1}{2}
$$


````python
import numpy as np
import matplotlib.pyplot as plt

# Verilen nokta ve vektr
A = np.array([1, 2])  # Nokta A(1,2)
v = np.array([2, 3])  # Yn vektr [2,3]

# Eim hesaplama (v'nin y ve x bileenlerinin oran)
m = v[1] / v[0]

# Dorunun denklemi: y - y1 = m(x - x1)
# y = m(x - x1) + y1
def line_eq(x):
    return m * (x - A[0]) + A[1]

# X ekseninde izim yapabileceimiz deerler
x_vals = np.linspace(-10, 10, 400)
y_vals = line_eq(x_vals)

# izim
plt.plot(x_vals, y_vals, label="y = (3/2)x + (1/2)")

# Nokta A(1,2)'yi ekleyelim
plt.scatter(A[0], A[1], color='red', label='Point A(1,2)', zorder=5)

# Grafik zellikleri
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.legend()

# Balk ve gsterim
plt.title("Equation of the Line Passing Through A(1,2) and Parallel to v = [2,3]")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## Problem: Parallel and Perpendicular Lines

**Given**:  
We are given the line \( y = 2x + 3 \).

**Find**:  
An example of a line that is:
1. Parallel to \( y = 2x + 3 \)
2. Perpendicular to \( y = 2x + 3 \)

---

## Solution

### 1. **Slope of the Given Line**

The given line is \( y = 2x + 3 \), which is in slope-intercept form \( y = mx + b \). Here:
- \( m = 2 \) (slope of the given line)

### 2. **Parallel Line**

A line parallel to \( y = 2x + 3 \) will have the same slope (\( m = 2 \)) but a different y-intercept.  
For example:

$$
y = 2x + 1
$$

### 3. **Perpendicular Line**

A line perpendicular to \( y = 2x + 3 \) will have a slope that is the negative reciprocal of \( m = 2 \).  
The negative reciprocal is:

$$
m = -\frac{1}{2}
$$

An example of such a line is:

$$
y = -\frac{1}{2}x + 4
$$

---

### Final Answer

- **Parallel Line**:  
  $$ y = 2x + 1 $$

- **Perpendicular Line**:  
  $$ y = -\frac{1}{2}x + 4 $$


````python
import matplotlib.pyplot as plt
import numpy as np

# Define the original line y = 2x + 3
def original_line(x):
    return 2 * x + 3

# Define a parallel line y = 2x + 1
def parallel_line(x):
    return 2 * x + 1

# Define a perpendicular line y = -1/2 * x + 4
def perpendicular_line(x):
    return -0.5 * x + 4

# Generate x values
x = np.linspace(-10, 10, 400)

# Plot the lines
plt.figure(figsize=(8, 6))
plt.plot(x, original_line(x), label="Original Line: $y = 2x + 3$", color="blue")
plt.plot(x, parallel_line(x), label="Parallel Line: $y = 2x + 1$", color="green")
plt.plot(x, perpendicular_line(x), label="Perpendicular Line: $y = -\\frac{1}{2}x + 4$", color="red")

# Add labels and legend
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title("Lines: Original, Parallel, and Perpendicular")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
````

## Problem: Distance from a Point to a Line

**Given**:  
We have the line \( y = 2x + 3 \) and the point \( A(1, 2) \).

**Find**:  
The distance from point \( A \) to the line \( y = 2x + 3 \).

---

## Solution

### 1. **Standard Form of the Line**

The given line is in slope-intercept form \( y = mx + b \). To find the distance, we need the standard form \( Ax + By + C = 0 \).  

From \( y = 2x + 3 \), we rewrite it as:

$$
2x - y + 3 = 0
$$

Here:
- \( A = 2 \),
- \( B = -1 \),
- \( C = 3 \).

---

### 2. **Distance Formula**

The distance \( d \) from a point \( (x_1, y_1) \) to a line \( Ax + By + C = 0 \) is given by:

$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

Substitute \( A = 2 \), \( B = -1 \), \( C = 3 \), and the point \( A(1, 2) \), where \( x_1 = 1 \) and \( y_1 = 2 \).

---

### 3. **Substitute Values**

Substitute into the formula:

$$
d = \frac{|2(1) + (-1)(2) + 3|}{\sqrt{2^2 + (-1)^2}}
$$

Simplify the numerator:

$$
d = \frac{|2 - 2 + 3|}{\sqrt{4 + 1}}
$$

$$
d = \frac{|3|}{\sqrt{5}}
$$

---

### 4. **Simplify**

The distance is:

$$
d = \frac{3}{\sqrt{5}}
$$

To rationalize the denominator:

$$
d = \frac{3\sqrt{5}}{5}
$$

---

### Final Answer

The distance from point \( A(1, 2) \) to the line \( y = 2x + 3 \) is:

$$
\frac{3\sqrt{5}}{5}
$$


````python
import numpy as np
import matplotlib.pyplot as plt

# Define the line y = 2x + 3
def line_equation(x):
    return 2 * x + 3

# Define the point A(1, 2)
point = (1, 2)

# Line coefficients in standard form: 2x - y + 3 = 0
A, B, C = 2, -1, 3

# Distance formula
def distance_from_point_to_line(x1, y1, A, B, C):
    numerator = abs(A * x1 + B * y1 + C)
    denominator = np.sqrt(A**2 + B**2)
    return numerator / denominator

# Calculate the distance
distance = distance_from_point_to_line(point[0], point[1], A, B, C)

# Visualization
x = np.linspace(-10, 10, 400)
y = line_equation(x)

# Perpendicular line (passes through A and is perpendicular to y=2x+3)
slope_perpendicular = -1 / 2
y_perpendicular = lambda x: slope_perpendicular * (x - point[0]) + point[1]

# Generate perpendicular line data
x_perpendicular = np.linspace(-10, 10, 400)
y_perpendicular_values = y_perpendicular(x_perpendicular)

# Plot the main line and the point
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Line: $y = 2x + 3$", color="blue")
plt.scatter(*point, color="red", label="Point A(1, 2)")
plt.plot(x_perpendicular, y_perpendicular_values, label="Perpendicular Line", linestyle="--", color="green")

# Annotate distance
plt.plot([point[0], point[0] + distance * np.sqrt(1 / (1 + 2**2))],
         [point[1], point[1] - distance * (2 / np.sqrt(1 + 2**2))],
         color="orange", linestyle=":", label=f"Distance = {distance:.2f}")

# Add labels and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Distance from a Point to a Line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
````

# Problem: Find the Equation of the Line

The line intersects the coordinate axes at points \( A(2, 0) \) and \( B(0, 3) \). Find the equation of the line.

---

## Solution

### 1. Find the Slope
The slope \( m \) of the line passing through two points \( (x_1, y_1) \) and \( (x_2, y_2) \) is given by the formula:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substituting the given points \( A(2, 0) \) and \( B(0, 3) \):

$$
m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -\frac{3}{2}
$$

### 2. Find the y-intercept (\( b \))
The equation of the line is:

$$
y = mx + b
$$

We already know the slope \( m = -\frac{3}{2} \). To find \( b \), substitute one of the points (e.g., \( A(2, 0) \)) into the equation:

$$
0 = -\frac{3}{2}(2) + b
$$

$$
0 = -3 + b
$$

$$
b = 3
$$

### 3. Write the Equation of the Line
Substitute \( m = -\frac{3}{2} \) and \( b = 3 \) into the equation of the line:

$$
y = -\frac{3}{2}x + 3
$$

---

### Final Answer
The equation of the line is:

$$
y = -\frac{3}{2}x + 3
$$


````python
import matplotlib.pyplot as plt
import numpy as np

# Given points A(2, 0) and B(0, 3)
x1, y1 = 2, 0
x2, y2 = 0, 3

# Calculate the slope (m)
m = (y2 - y1) / (x2 - x1)

# Calculate the y-intercept (b)
b = y1 - m * x1

# Generate x values for the graph
x = np.linspace(-1, 3, 100)

# Equation of the line: y = mx + b
y = m * x + b

# Plot the line
plt.plot(x, y, label=f"y = {m}x + {b}", color='blue')

# Plot the points A(2, 0) and B(0, 3)
plt.scatter([x1, x2], [y1, y2], color="red", zorder=5)

# Labels and title
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Equation of the Line")
plt.legend()

# Show the plot
plt.show()
````

## Problem: Angle Between a Line and the \( x \)-Axis

**Given**:  
The line \( y = x + 3 \).

**Find**:  
The angle between the line \( y = x + 3 \) and the \( x \)-axis.

---

## Solution

### 1. **Slope of the Line**

The given line is in the slope-intercept form \( y = mx + b \), where \( m \) is the slope of the line.  
Here, \( m = 1 \).

---

### 2. **Angle Formula**

The angle \( \theta \) between a line and the \( x \)-axis is given by:

$$
\theta = \arctan(m)
$$

Where \( m \) is the slope of the line.

---

### 3. **Substitute the Slope**

Substitute \( m = 1 \) into the formula:

$$
\theta = \arctan(1)
$$

---

### 4. **Calculate the Angle**

The value of \( \arctan(1) \) is:

$$
\theta = 45^\circ
$$

---

### Final Answer

The angle between the line \( y = x + 3 \) and the \( x \)-axis is:

$$
45^\circ
$$


````python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define the line y = x + 3
def line_equation(x):
    return x + 3

# Generate x values
x = np.linspace(-10, 10, 400)
y = line_equation(x)

# Calculate the angle with the x-axis
m = 1  # Slope of the line
theta_radians = math.atan(m)
theta_degrees = math.degrees(theta_radians)

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Line: $y = x + 3$", color="blue")

# Plot the x-axis
plt.axhline(0, color='black', linewidth=0.5, label="x-axis")

# Annotate the angle
plt.text(2, 4, f"$\\theta = {theta_degrees:.2f}^\\circ$", fontsize=12, color="red")

# Add grid, labels, and legend
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Angle Between the Line and the x-Axis")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
````

## Problem: Perpendicular Vector to a Line

**Given**:  
The line equation \( x + y + 1 = 0 \).

**Find**:  
A vector perpendicular to the given line.

---

## Solution

### 1. **General Form of a Line**

The general form of a line is:

$$
ax + by + c = 0
$$

Here:
- \( a = 1 \),
- \( b = 1 \),
- \( c = 1 \).

---

### 2. **Perpendicular Vector**

A vector perpendicular to the line has the components \( [a, b] \), where \( a \) and \( b \) are the coefficients of \( x \) and \( y \) in the line equation.

Thus, a perpendicular vector is:

$$
\vec{v} = [1, 1]
$$

---

### Final Answer

A vector perpendicular to the line \( x + y + 1 = 0 \) is:

$$
\vec{v} = [1, 1]
$$


````python
import numpy as np
import matplotlib.pyplot as plt

# Line equation: x + y + 1 = 0
def line_equation(x):
    return -x - 1

# Perpendicular vector
vector = [1, 1]  # [a, b] from the line equation

# Generate x values for the line
x = np.linspace(-10, 10, 400)
y = line_equation(x)

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Line: $x + y + 1 = 0$", color="blue")

# Plot the perpendicular vector starting from the origin
origin = [0, 0]
plt.quiver(origin[0], origin[1], vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color="red", label="Perpendicular Vector")

# Add grid, labels, and legend
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Line and Perpendicular Vector")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.axis('equal')  # Equal scaling for x and y axes
plt.show()
````

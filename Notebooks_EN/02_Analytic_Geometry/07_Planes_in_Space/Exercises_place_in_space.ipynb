### Problem: Find the Equation of the Plane

Given three points \( A(1, 2, 3) \), \( B(3, 4, 5) \), and \( C(2, 1, 4) \), find the equation of the plane passing through these points.

---

### Solution:

To find the equation of the plane, we can use the general formula for the equation of a plane:

$$
Ax + By + Cz + D = 0
$$

Where \( A \), \( B \), and \( C \) are the components of a normal vector to the plane, and \( D \) is a constant.

#### Step 1: Find Two Vectors in the Plane
We can find two vectors that lie in the plane by subtracting the coordinates of the points:

- Vector \( \mathbf{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2) \)
- Vector \( \mathbf{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1) \)

#### Step 2: Find the Normal Vector to the Plane
The normal vector \( \mathbf{N} \) to the plane is the cross product of the vectors \( \mathbf{AB} \) and \( \mathbf{AC} \):

$$
\mathbf{N} = \mathbf{AB} \times \mathbf{AC}
$$

We can calculate the cross product using the determinant of a matrix:

$$
\mathbf{N} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 2 & 2 & 2 \\ 1 & -1 & 1 \end{vmatrix}
$$

#### Step 3: Find the Equation of the Plane
Once we have the normal vector, we can plug it into the plane equation \( Ax + By + Cz + D = 0 \), using point \( A(1, 2, 3) \) to solve for \( D \).

---

### Final Answer:
The equation of the plane is:

$$
Ax + By + Cz + D = 0
$$

where \( A \), \( B \), and \( C \) are the components of the normal vector, and \( D \) is calculated from the given points.


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Verilen noktalar A(1, 2, 3), B(3, 4, 5), C(2, 1, 4)
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Adım 1: AB ve AC vektörlerini bulun
AB = B - A
AC = C - A

# Adım 2: Normal vektörü bulun (AB ve AC'nin çapraz çarpımı)
# np.cross fonksiyonu iki vektörün çapraz çarpımını hesaplar.
N = np.cross(AB, AC)

# Adım 3: Düzlemin denklemi Ax + By + Cz + D = 0
# Normal vektör N bize A, B, C değerlerini verir
a, b, c = N

# Adım 4: A noktasını denklemde yerine koyarak D'yi bulun
# np.dot fonksiyonu iki vektörün nokta çarpımını hesaplar.
D = -np.dot(N, A)

# Düzlemin denklemini yazdırın
print(f"Düzlemin denklemi: {a}x + {b}y + {c}z + ({D}) = 0")

# Adım 5: Noktaları ve düzlemi görselleştirin
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# A, B, C noktalarını çizdirin
ax.scatter(A[0], A[1], A[2], color='r', label='A(1,2,3)')
ax.scatter(B[0], B[1], B[2], color='g', label='B(3,4,5)')
ax.scatter(C[0], C[1], C[2], color='b', label='C(2,1,4)')

# Düzlem için bir ızgara oluşturun
xx, yy = np.meshgrid(np.linspace(-2, 4, 20), np.linspace(-2, 4, 20))

# Düzlem denklemini kullanarak z değerlerini hesaplayın
zz = (-a * xx - b * yy - D) / c

# Düzlemi çizdirin
ax.plot_surface(xx, yy, zz, alpha=0.5, rstride=5, cstride=5)

# Eksen etiketlerini ayarlayın
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Eksen sınırlarını ayarlayın (isteğe bağlı)
ax.set_xlim([-2, 4])
ax.set_ylim([-2, 4])
ax.set_zlim([0, 6])

# Göstergeyi gösterin
ax.legend()

# Çizimi gösterin
plt.show()
````

# Problem: Find the Equation of the Plane

The plane passes through the point \( A(1, 2, 3) \) and is parallel to the plane \( 2x + 3y + 4z = 5 \). Find the equation of the plane.

---

## Solution

### 1. Understanding the Problem
- The given plane is parallel to the plane \( 2x + 3y + 4z = 5 \).
- Parallel planes have the same normal vector.

### 2. Normal Vector of the Given Plane
The equation of the plane is \( 2x + 3y + 4z = 5 \). The normal vector to this plane is:

$$ \mathbf{n} = (2, 3, 4) $$

Since the desired plane is parallel to this one, it will have the same normal vector \( \mathbf{n} = (2, 3, 4) \).

### 3. General Equation of the Plane
The general equation of a plane is:

$$ ax + by + cz = d $$

Where \( a, b, c \) are the components of the normal vector to the plane. For our case, \( a = 2, b = 3, c = 4 \).

Thus, the equation of the plane becomes:

$$ 2x + 3y + 4z = d $$

### 4. Find \( d \) Using the Point \( A(1, 2, 3) \)
Since the plane passes through the point \( A(1, 2, 3) \), substitute \( x = 1 \), \( y = 2 \), and \( z = 3 \) into the equation:

$$ 2(1) + 3(2) + 4(3) = d $$

$$ 2 + 6 + 12 = d $$

$$ d = 20 $$

### 5. Final Equation of the Plane
The equation of the plane is:

$$ 2x + 3y + 4z = 20 $$

---

### Final Answer
The equation of the plane is:

$$ 2x + 3y + 4z = 20 $$



````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Equation of the first plane: 2x + 3y + 4z = 5
# Normal vector: (2, 3, 4)

# Create a grid of x and y values
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Calculate z values for the first plane
Z1 = (5 - 2*X - 3*Y) / 4

# Equation of the parallel plane: 2x + 3y + 4z = d
# Find d using point A(1, 2, 3)
d = 2*1 + 3*2 + 4*3  # d = 20

# Calculate z values for the parallel plane
Z2 = (d - 2*X - 3*Y) / 4

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the first plane
ax.plot_surface(X, Y, Z1, cmap='viridis', edgecolor='k', alpha=0.5, label='First Plane: 2x + 3y + 4z = 5')

# Plot the parallel plane
ax.plot_surface(X, Y, Z2, cmap='magma', edgecolor='b', alpha=0.5, label='Parallel Plane: 2x + 3y + 4z = 20')

# Plot point A
ax.scatter(1, 2, 3, color='r', s=100, label="Point A(1, 2, 3)")

# Set axis labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Parallel Planes')

# Display the legend
plt.legend()

# Show the plot
plt.show()
````

# Problem: Find the Equation of the Plane

The plane passes through the point \( A(1, 2, 3) \) and is perpendicular to the normal vector \( \mathbf{n} = [2, 3, 4] \). Find the equation of the plane.

---

## Solution

### 1. Equation of the Plane
The general equation of a plane is given by:

$$
a(x - x_0) + b(y - y_0) + c(z - z_0) = 0
$$

where \( (x_0, y_0, z_0) \) is a point on the plane, and \( [a, b, c] \) is the normal vector to the plane.

### 2. Substitute the Known Values
We are given:
- The point \( A(1, 2, 3) \) is on the plane, so \( x_0 = 1 \), \( y_0 = 2 \), and \( z_0 = 3 \).
- The normal vector \( \mathbf{n} = [2, 3, 4] \), so \( a = 2 \), \( b = 3 \), and \( c = 4 \).

Substitute these values into the equation of the plane:

$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$

### 3. Simplify the Equation
Now, simplify the equation:

$$
2x - 2 + 3y - 6 + 4z - 12 = 0
$$

$$
2x + 3y + 4z - 20 = 0
$$

### 4. Final Answer
The equation of the plane is:

$$
2x + 3y + 4z = 20
$$


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given point A and normal vector n
A = np.array([1, 2, 3])
n = np.array([2, 3, 4])

# Equation of the plane: n . (X - A) = 0
# We need to plot the plane with the equation: 2x + 3y + 4z = 20

# Create a meshgrid for plotting the plane
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# Solve for Z from the plane equation: 2x + 3y + 4z = 20 => z = (20 - 2x - 3y) / 4
Z = (20 - 2*X - 3*Y) / 4

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100)

# Plot point A
ax.scatter(A[0], A[1], A[2], color="r", s=100, label="Point A(1, 2, 3)")

# Plot the normal vector
ax.quiver(A[0], A[1], A[2], n[0], n[1], n[2], color="b", length=3, normalize=True, label="Normal Vector")

# Labels and Title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("Plane Equation: 2x + 3y + 4z = 20")

# Show plot
plt.legend()
plt.show()
````

### Problem: Find the Line of Intersection of Two Planes

Given two planes:  
1. \( 2x + 3y + 4z = 5 \)  
2. \( 3x + 4y + 2z = 6 \)  

Find the equation of the line of intersection of these planes.

---

### Solution:

The line of intersection of two planes can be found parametrically as follows:

#### Step 1: Represent the Planes
The equations of the planes are:  
\[
2x + 3y + 4z = 5
\]  
\[
3x + 4y + 2z = 6
\]

#### Step 2: Parametrize the Line
To find the intersection, we express the system in terms of a parameter \( t \):  
Let \( z = t \).

Substitute \( z = t \) into both plane equations:

1. \( 2x + 3y + 4t = 5 \)  
   Rearrange:  
   \[
   2x + 3y = 5 - 4t
   \] \tag{1}

2. \( 3x + 4y + 2t = 6 \)  
   Rearrange:  
   \[
   3x + 4y = 6 - 2t
   \] \tag{2}

#### Step 3: Solve for \( x \) and \( y \)
From equations (1) and (2), we solve for \( x \) and \( y \) in terms of \( t \):  

1. Multiply equation (1) by 3 and equation (2) by 2 to eliminate \( x \):  
   \[
   6x + 9y = 15 - 12t
   \]  
   \[
   6x + 8y = 12 - 4t
   \]

   Subtract the second equation from the first:  
   \[
   y = 3 - 8t
   \]

2. Substitute \( y = 3 - 8t \) into equation (1):  
   \[
   2x + 3(3 - 8t) = 5 - 4t
   \]  
   \[
   2x + 9 - 24t = 5 - 4t
   \]  
   \[
   2x = -4 + 20t
   \]  
   \[
   x = -2 + 10t
   \]

#### Step 4: Parametric Equations of the Line
The line of intersection is given by the parametric equations:  
\[
x = -2 + 10t
\]  
\[
y = 3 - 8t
\]  
\[
z = t
\]

---

### Final Answer:
The line of intersection is:  
\[
(x, y, z) = (-2 + 10t, 3 - 8t, t), \quad t \in \mathbb{R}
\]


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the equations of the planes
# 2x + 3y + 4z = 5
# 3x + 4y + 2z = 6

# Create a grid of x and y values using numpy.linspace and numpy.meshgrid.
# This grid will be used to plot the surfaces of the planes.
x = np.linspace(-5, 5, 10)  # Create 10 points between -5 and 5 for x
y = np.linspace(-5, 5, 10)  # Create 10 points between -5 and 5 for y
X, Y = np.meshgrid(x, y)     # Create a grid of x and y values

# Calculate the z values for each plane using the plane equations.
# We solve for z in terms of x and y.
Z1 = (5 - 2*X - 3*Y) / 4  # Calculate z values for the first plane
Z2 = (6 - 3*X - 4*Y) / 2  # Calculate z values for the second plane

# Define the parametric equations for the line of intersection.
# We found these equations in the previous steps.
t = np.linspace(-2, 2, 100)  # Create 100 points between -2 and 2 for the parameter t
x_line = -2 + 10*t          # Calculate x values for the line
y_line = 3 - 8*t           # Calculate y values for the line
z_line = t                  # Calculate z values for the line

# Create a 3D plot using matplotlib.pyplot.figure and add a subplot.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Plot the surfaces of the planes using ax.plot_surface.
# alpha controls the transparency of the surfaces.
ax.plot_surface(X, Y, Z1, alpha=0.5, label='2x + 3y + 4z = 5')  # Plot the first plane
ax.plot_surface(X, Y, Z2, alpha=0.5, label='3x + 4y + 2z = 6')  # Plot the second plane

# Plot the line of intersection using ax.plot.
ax.plot(x_line, y_line, z_line, color='red', label='Intersection Line')

# Set the axis labels using ax.set_xlabel, ax.set_ylabel, and ax.set_zlabel.
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend using ax.legend.
ax.legend()

# Show the plot using plt.show.
plt.show()
````

### Problem: Find the Equation of the Plane

Given:  
- A point \( A(1, 2, 3) \)  
- Two vectors parallel to the plane:  
  \[
  \mathbf{v_1} = [1, 0, 1], \quad \mathbf{v_2} = [0, 1, -1]
  \]

Find the equation of the plane passing through the point and parallel to the given vectors.

---

### Solution:

The equation of a plane can be written as:  
\[
Ax + By + Cz + D = 0
\]
where \( [A, B, C] \) is the normal vector to the plane.

#### Step 1: Find the Normal Vector  
The normal vector \( \mathbf{n} \) to the plane is perpendicular to both \( \mathbf{v_1} \) and \( \mathbf{v_2} \). We find \( \mathbf{n} \) by taking the cross product of \( \mathbf{v_1} \) and \( \mathbf{v_2} \):  

\[
\mathbf{n} = \mathbf{v_1} \times \mathbf{v_2}
\]

Using the determinant formula:  
\[
\mathbf{n} = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 0 & 1 \\
0 & 1 & -1
\end{vmatrix}
\]

Expand the determinant:  
\[
\mathbf{n} = \hat{i} \begin{vmatrix} 0 & 1 \\ 1 & -1 \end{vmatrix}
- \hat{j} \begin{vmatrix} 1 & 1 \\ 0 & -1 \end{vmatrix}
+ \hat{k} \begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix}
\]

\[
\mathbf{n} = \hat{i}(0 \cdot -1 - 1 \cdot 1) - \hat{j}(1 \cdot -1 - 1 \cdot 0) + \hat{k}(1 \cdot 1 - 0 \cdot 0)
\]

\[
\mathbf{n} = \hat{i}(-1) - \hat{j}(-1) + \hat{k}(1)
\]

\[
\mathbf{n} = [-1, 1, 1]
\]

Thus, the normal vector is:  
\[
\mathbf{n} = [-1, 1, 1]
\]

#### Step 2: Use the Point-Normal Form of the Plane  
The equation of the plane passing through \( A(1, 2, 3) \) and perpendicular to \( \mathbf{n} = [-1, 1, 1] \) is given by:  
\[
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0
\]

Simplify:  
\[
-x + y + z - 1 - 2 - 3 = 0
\]

\[
-x + y + z - 6 = 0
\]

\[
x - y - z + 6 = 0
\]

---

### Final Answer:
The equation of the plane is:  
\[
x - y - z + 6 = 0
\]


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Nokta ve vektörler
A = np.array([1, 2, 3])
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])

# Normal vektörün hesaplanması
normal_vector = np.cross(v1, v2)

# Düzlem denklemi: -x + y + z = 4
# Düzlem denklemi ax + by + cz + d = 0 şeklinde yazılabilir.
a, b, c = normal_vector
d = -np.dot(normal_vector, A)

# Meshgrid ile düzlem oluşturma
xx, yy = np.meshgrid(range(-10, 10), range(-10, 10))
zz = (-d - a * xx - b * yy) / c

# 3D görselleştirme
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, zz, alpha=0.5, rstride=100, cstride=100)

# Noktayı ve vektörleri çizdirme
ax.scatter(A[0], A[1], A[2], color='red', s=100, label='Point A (1,2,3)')  # Nokta
ax.quiver(A[0], A[1], A[2], v1[0], v1[1], v1[2], color='blue', length=3, normalize=True, label='Vector v1 [1,0,1]')
ax.quiver(A[0], A[1], A[2], v2[0], v2[1], v2[2], color='green', length=3, normalize=True, label='Vector v2 [0,1,-1]')

# Ekseni etiketleme
ax.set_xlabel('X ekseni')
ax.set_ylabel('Y ekseni')
ax.set_zlabel('Z ekseni')

# Grafik başlığı
ax.set_title('Düzlem ve Paralel Vektörler')
ax.legend()

plt.show()
````

### Problem: Find Planes Parallel and Perpendicular to \( 2x + 3y + 4z = 5 \)

Given the plane:  
\[
2x + 3y + 4z = 5
\]

Find:  
1. A plane parallel to it.  
2. A plane perpendicular to it.

---

### Solution:

#### 1. Parallel Plane
Two planes are parallel if they share the same normal vector. The normal vector for the given plane is:  
\[
\mathbf{n} = [2, 3, 4]
\]

To find a parallel plane, we keep the same coefficients for \( x \), \( y \), and \( z \), but change the constant term.  
For example, the plane:  
\[
2x + 3y + 4z = 10
\]
is parallel to \( 2x + 3y + 4z = 5 \).

---

#### 2. Perpendicular Plane
Two planes are perpendicular if the dot product of their normal vectors is zero.  

Let the equation of the perpendicular plane be:  
\[
Ax + By + Cz + D = 0
\]
where \( \mathbf{n}_\text{new} = [A, B, C] \) is the normal vector.  

The dot product of the normal vectors must satisfy:  
\[
\mathbf{n} \cdot \mathbf{n}_\text{new} = 0
\]

Substituting \( \mathbf{n} = [2, 3, 4] \):  
\[
2A + 3B + 4C = 0
\]

Choose \( A = 3 \), \( B = -2 \), and \( C = 0 \) as a solution (one possible choice):  
\[
3(2) + (-2)(3) + 0(4) = 0
\]

Thus, the equation of the perpendicular plane can be:  
\[
3x - 2y + D = 0
\]

Using any value for \( D \), for instance \( D = 5 \):  
\[
3x - 2y + 5 = 0
\]

---

### Final Answer:
1. A parallel plane:  
   \[
   2x + 3y + 4z = 10
   \]

2. A perpendicular plane:  
   \[
   3x - 2y + 5 = 0
   \]


````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of points in the X-Y plane
xx, yy = np.meshgrid(range(-10, 10), range(-10, 10))

# Define the planes
zz1 = (5 - 2 * xx - 3 * yy) / 4  # Original plane: 2x + 3y + 4z = 5
zz2 = (10 - 2 * xx - 3 * yy) / 4  # Parallel plane: 2x + 3y + 4z = 10
zz3 = (4 * yy) / 3  # Perpendicular plane: 4y - 3z = 0

# Plot the planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original plane
ax.plot_surface(xx, yy, zz1, alpha=0.5, rstride=100, cstride=100, color='blue', label='2x + 3y + 4z = 5')

# Plot the parallel plane
ax.plot_surface(xx, yy, zz2, alpha=0.5, rstride=100, cstride=100, color='green', label='2x + 3y + 4z = 10')

# Plot the perpendicular plane
ax.plot_surface(xx, yy, zz3, alpha=0.5, rstride=100, cstride=100, color='red', label='4y - 3z = 0')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Planes Visualization')

# Show legend
ax.legend()

# Show the plot
plt.show()
````

### Plane Equation and Distance from a Point to a Plane

In this section, we will calculate the distance from the point \(A(1, 2, 3)\) to the plane \(2x + 3y + 4z = 5\).

#### Distance Formula

To find the distance from a point \((x_1, y_1, z_1)\) to a plane given by \(ax + by + cz + d = 0\), we use the following formula:

$$ \text{Distance} = \frac{|ax_1 + by_1 + cz_1 + d|}{\sqrt{a^2 + b^2 + c^2}} $$

Given plane: \(2x + 3y + 4z = 5\)

We can rewrite this plane equation in the form \(2x + 3y + 4z - 5 = 0\), where \(a = 2\), \(b = 3\), \(c = 4\), and \(d = -5\).

Point: \(A(1, 2, 3)\)

#### Calculating the Distance

To calculate the distance from the point to the plane, we substitute the values into the formula:

$$ \text{Distance} = \frac{|2 \cdot 1 + 3 \cdot 2 + 4 \cdot 3 - 5|}{\sqrt{2^2 + 3^2 + 4^2}} $$

Calculating the numerator:

$$ |2 \cdot 1 + 3 \cdot 2 + 4 \cdot 3 - 5| = |2 + 6 + 12 - 5| = |15| = 15 $$

Calculating the denominator:

$$ \sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29} $$

Therefore, the distance is:

$$ \text{Distance} = \frac{15}{\sqrt{29}} $$

This is the exact distance. The approximate value is:

$$ \text{Distance} \approx \frac{15}{\sqrt{29}} \approx 2.784 $$

Thus, the distance from the point \(A(1, 2, 3)\) to the plane \(2x + 3y + 4z = 5\) is approximately 2.784 units.

### Equation of the Plane Intersecting Coordinate Axes

The plane intersects the coordinate axes at points \(A(2, 0, 0)\), \(B(0, 3, 0)\), and \(C(0, 0, 4)\). To find the equation of the plane, we can use the intercept form of the plane equation.

#### Intercept Form of the Plane Equation

The intercept form of the plane equation is given by:

$$ \frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1 $$

where \(a\), \(b\), and \(c\) are the intercepts of the plane with the x-axis, y-axis, and z-axis, respectively.

Given points:
- Point \(A(2, 0, 0)\) gives us \(a = 2\)
- Point \(B(0, 3, 0)\) gives us \(b = 3\)
- Point \(C(0, 0, 4)\) gives us \(c = 4\)

Substituting these values into the intercept form, we get:

$$ \frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1 $$

To express this equation in the standard form \(ax + by + cz = d\), multiply through by the least common multiple of the denominators (2, 3, and 4), which is 12:

$$ 12 \left( \frac{x}{2} + \frac{y}{3} + \frac{z}{4} \right) = 12 \cdot 1 $$

Simplifying, we get:

$$ 6x + 4y + 3z = 12 $$

Thus, the equation of the plane is:

$$ 6x + 4y + 3z = 12 $$

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the points A, B, and C
A = np.array([2, 0, 0])
B = np.array([0, 3, 0])
C = np.array([0, 0, 4])

# Calculate normal vector of the plane
normal = np.cross(B - A, C - A)

# Plane equation coefficients
a, b, c = normal
d = -np.dot(normal, A)

# Define the plane
xx, yy = np.meshgrid(range(-1, 4), range(-1, 4))
zz = (-a * xx - b * yy - d) / c

# Plot the plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, zz, alpha=0.5, rstride=100, cstride=100, color='blue')

# Plot the points A, B, and C
ax.scatter(A[0], A[1], A[2], color='red', s=100, label='Point A (2, 0, 0)')
ax.scatter(B[0], B[1], B[2], color='green', s=100, label='Point B (0, 3, 0)')
ax.scatter(C[0], C[1], C[2], color='purple', s=100, label='Point C (0, 0, 4)')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Plane and Points Visualization')

# Show legend
ax.legend()

# Show the plot
plt.show()
````

### Angle Between Two Planes

To calculate the angle between the plane \(x + y + z = 1\) and the plane \(x = 0\) (i.e., the \(yz\) plane), we can use the dot product of their normal vectors.

#### Formula for the Angle Between Two Planes

The formula for the angle \(\theta\) between two planes with normal vectors \(\mathbf{n_1}\) and \(\mathbf{n_2}\) is given by:

$$ \cos(\theta) = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|} $$

where \(\mathbf{n_1}\) and \(\mathbf{n_2}\) are the normal vectors of the two planes.

#### Normal Vectors

- For the plane \(x + y + z = 1\), the normal vector is \(\mathbf{n_1} = (1, 1, 1)\).
- For the plane \(x = 0\) (the \(yz\) plane), the normal vector is \(\mathbf{n_2} = (1, 0, 0)\).

#### Calculating the Angle

The dot product \(\mathbf{n_1} \cdot \mathbf{n_2}\) is:

$$ \mathbf{n_1} \cdot \mathbf{n_2} = 1 \cdot 1 + 1 \cdot 0 + 1 \cdot 0 = 1 $$

The magnitudes \(|\mathbf{n_1}|\) and \(|\mathbf{n_2}|\) are:

$$ |\mathbf{n_1}| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3} $$
$$ |\mathbf{n_2}| = \sqrt{1^2 + 0^2 + 0^2} = 1 $$

Therefore, the cosine of the angle \(\theta\) is:

$$ \cos(\theta) = \frac{1}{\sqrt{3} \cdot 1} = \frac{1}{\sqrt{3}} $$

Thus, the angle \(\theta\) is:

$$ \theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right) $$

### Python Code to Calculate and Visualize the Angle

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vectors

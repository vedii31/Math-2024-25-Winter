# Equation of a Sphere

In this section, we will derive the equation of a sphere with a given center and radius.

## Step 1: General Equation of a Sphere

The general equation of a sphere is given by:

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Here, \((x_0, y_0, z_0)\) are the coordinates of the center of the sphere and \(r\) is the radius of the sphere.

## Step 2: Substituting the Given Information

Given the center of the sphere \( P = (1, 2, 3) \) and the radius \( r = 3 \), we substitute these values into the general equation:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$

## Step 3: Simplifying the Equation

Next, we simplify the equation:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

This is the equation of a sphere with the center at \( P = (1, 2, 3) \) and a radius of \( r = 3 \).

## Summary

The equation of a sphere with the center at \( P = (1, 2, 3) \) and a radius of \( r = 3 \) is:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Center and radius of the sphere
center = np.array([1, 2, 3])
radius = 3

# Create a grid of points
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the sphere
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, color='b', alpha=0.6, rstride=5, cstride=5, edgecolor='k')

# Plot the center
ax.scatter(center[0], center[1], center[2], color='r', s=100)

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Sphere with center at (1, 2, 3) and radius 3')

plt.show()
```

# Common Points Between Two Spheres

To determine if the spheres with equations \( x^2 + y^2 + z^2 = 1 \) and \( x^2 + y^2 + z^2 = 2 \) have any common points, we analyze their equations.

## Step-by-step Analysis

1. **Equations of the Spheres**:
   - Sphere 1: \( x^2 + y^2 + z^2 = 1 \)
   - Sphere 2: \( x^2 + y^2 + z^2 = 2 \)

2. **Comparison of the Equations**:
   - The left-hand side of both equations represents the same expression \( x^2 + y^2 + z^2 \).

3. **Checking for Common Points**:
   - If a point \((x, y, z)\) lies on both spheres, it must satisfy both equations simultaneously.
   - However, for a single point \((x, y, z)\), \( x^2 + y^2 + z^2 \) cannot be equal to both 1 and 2 at the same time.

## Conclusion
Since the values on the right-hand side of the equations are different, there are no points \((x, y, z)\) that can satisfy both equations simultaneously. Therefore, the spheres do not have any common points.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the radii and centers of the spheres
radius1 = 1
center1 = np.array([0, 0, 0])  # Center at origin

radius2 = np.sqrt(2)
center2 = np.array([0, 0, 0])  # Center at origin

# Create a grid of points for the first sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x1 = center1[0] + radius1 * np.outer(np.cos(u), np.sin(v))
y1 = center1[1] + radius1 * np.outer(np.sin(u), np.sin(v))
z1 = center1[2] + radius1 * np.outer(np.ones(np.size(u)), np.cos(v))

# Create a grid of points for the second sphere
x2 = center2[0] + radius2 * np.outer(np.cos(u), np.sin(v))
y2 = center2[1] + radius2 * np.outer(np.sin(u), np.sin(v))
z2 = center2[2] + radius2 * np.outer(np.ones(np.size(u)), np.cos(v))

# Plot the spheres
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the first sphere
ax.plot_surface(x1, y1, z1, color='b', alpha=0.5, rstride=5, cstride=5, edgecolor='k', label='Sphere 1')

# Plot the second sphere
ax.plot_surface(x2, y2, z2, color='g', alpha=0.5, rstride=5, cstride=5, edgecolor='k', label='Sphere 2')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Visualization of Two Spheres')

# Add legend
ax.legend(['Sphere 1', 'Sphere 2'], loc='upper left')

# Show the plot
plt.show()

# Check if the spheres intersect
# Since both spheres are centered at the origin, they cannot intersect if their radii are different
if radius1 == radius2:
    print("The spheres intersect.")
else:
    print("The spheres do not intersect.")
```

## Intersection of Two Spheres

To find the curve formed by the intersection of the two spheres given by:

1. \( x^2 + y^2 + z^2 = 1 \)
2. \( (x - 1)^2 + y^2 + z^2 = 1 \)

we can follow these steps:

### Step 1: Expand the Equation of the Second Sphere

First, expand the equation of the second sphere:

$$
(x - 1)^2 + y^2 + z^2 = 1
$$

$$
(x - 1)^2 + y^2 + z^2 = 1 \Rightarrow x^2 - 2x + 1 + y^2 + z^2 = 1
$$

### Step 2: Simplify the Equation

Combine like terms:

$$
x^2 + y^2 + z^2 - 2x + 1 = 1
$$

Subtract 1 from both sides:

$$
x^2 + y^2 + z^2 - 2x = 0
$$

### Step 3: Use the Equation of the First Sphere

We know from the first sphere that:

$$
x^2 + y^2 + z^2 = 1
$$

Substitute \( x^2 + y^2 + z^2 = 1 \) into the simplified equation:

$$
1 - 2x = 0
$$

### Step 4: Solve for \( x \)

Solving for \( x \):

$$
1 = 2x \Rightarrow x = \frac{1}{2}
$$

### Step 5: Substitute \( x = \frac{1}{2} \) Back into the First Sphere Equation

Substitute \( x = \frac{1}{2} \) back into the equation of the first sphere:

$$
\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1
$$

$$
\frac{1}{4} + y^2 + z^2 = 1
$$

### Step 6: Simplify to Find the Equation of the Curve

Subtract \(\frac{1}{4}\) from both sides:

$$
y^2 + z^2 = 1 - \frac{1}{4}
$$

$$
y^2 + z^2 = \frac{3}{4}
$$

### Conclusion

The curve formed by the intersection of the two spheres is a circle in the plane \( x = \frac{1}{2} \) with radius \(\sqrt{\frac{3}{4}} = \frac{\sqrt{3}}{2} \). The equation of the curve is:

$$
x = \frac{1}{2}
$$

$$
y^2 + z^2 = \frac{3}{4}
$$

This explains the equation of the curve formed by the intersection of the two spheres and how it was derived.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Kürelerin merkezleri ve yarıçapları
center1 = np.array([0, 0, 0])
radius1 = 1

center2 = np.array([1, 0, 0])
radius2 = 1

# Kürelerin koordinatlarını oluşturmak için bir grid oluşturun
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

# İlk küre için koordinatlar
x1 = center1[0] + radius1 * np.outer(np.cos(u), np.sin(v))
y1 = center1[1] + radius1 * np.outer(np.sin(u), np.sin(v))
z1 = center1[2] + radius1 * np.outer(np.ones(np.size(u)), np.cos(v))

# İkinci küre için koordinatlar
x2 = center2[0] + radius2 * np.outer(np.cos(u), np.sin(v))
y2 = center2[1] + radius2 * np.outer(np.sin(u), np.sin(v))
z2 = center2[2] + radius2 * np.outer(np.ones(np.size(u)), np.cos(v))

# Kesişim çemberi için koordinatlar
theta = np.linspace(0, 2 * np.pi, 100)
x_intersect = 0.5 * np.ones_like(theta)
y_intersect = np.sqrt(0.75) * np.cos(theta)
z_intersect = np.sqrt(0.75) * np.sin(theta)

# Grafiği oluşturma
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# İlk küreyi çizme
ax.plot_surface(x1, y1, z1, color='b', alpha=0.5, rstride=5, cstride=5, edgecolor='k')

# İkinci küreyi çizme
ax.plot_surface(x2, y2, z2, color='g', alpha=0.5, rstride=5, cstride=5, edgecolor='k')

# Kesişim çemberini çizme
ax.plot(x_intersect, y_intersect, z_intersect, color='r', linewidth=2)

# Eksen etiketleri ve başlık
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Intersection of Two Spheres')

# Göster
plt.show()
```

## Finding the Equation of the Tangent Plane

To find the equation of the tangent plane to the paraboloid \( z = (x - 1)^2 + y^2 + 1 \) at the point \( P = (1, 0, 1) \), we need to follow these steps:

### Step 1: Verify the Point on the Paraboloid
First, check that the point \( P = (1, 0, 1) \) lies on the paraboloid:
$$ z = (x - 1)^2 + y^2 + 1 $$

Substitute \( x = 1 \) and \( y = 0 \) into the equation:
$$ z = (1 - 1)^2 + 0^2 + 1 = 1 $$

So, the point \( P = (1, 0, 1) \) is on the paraboloid.

### Step 2: Find the Partial Derivatives
The equation of the tangent plane at a point \( (x_0, y_0, z_0) \) on a surface \( z = f(x, y) \) is given by:
$$ z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0) $$

Calculate the partial derivatives \( f_x \) and \( f_y \):
$$ z = (x - 1)^2 + y^2 + 1 $$

$$ f(x, y) = (x - 1)^2 + y^2 + 1 $$

$$ f_x = \frac{\partial}{\partial x} [(x - 1)^2 + y^2 + 1] = 2(x - 1) $$

$$ f_y = \frac{\partial}{\partial y} [(x - 1)^2 + y^2 + 1] = 2y $$

Evaluate these at the point \( (1, 0) \):
$$ f_x(1, 0) = 2(1 - 1) = 0 $$

$$ f_y(1, 0) = 2(0) = 0 $$

### Step 3: Write the Equation of the Tangent Plane
Using the point \( (1, 0, 1) \) and the partial derivatives, the equation of the tangent plane is:
$$ z - 1 = 0(x - 1) + 0(y - 0) $$

Simplify the equation:
$$ z - 1 = 0 $$

$$ z = 1 $$

### Conclusion
The equation of the tangent plane to the paraboloid at the point \( P = (1, 0, 1) \) is \( z = 1 \).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the paraboloid function
def paraboloid(x, y):
    return (x - 1)**2 + y**2 + 1

# Create a grid of x and y values
x = np.linspace(-1, 3, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)
z = paraboloid(x, y)

# Define the tangent plane function
def tangent_plane(x, y):
    return 1  # The tangent plane at point (1, 0, 1) is z = 1

# Create a grid of x and y values for the tangent plane
x_tan = np.linspace(0, 2, 400)
y_tan = np.linspace(-1, 1, 400)
x_tan, y_tan = np.meshgrid(x_tan, y_tan)
z_tan = tangent_plane(x_tan, y_tan) * np.ones_like(x_tan)  # Ensure z_tan is a 2D array

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the paraboloid
ax.plot_surface(x, y, z, color='b', alpha=0.6, rstride=100, cstride=100, edgecolor='none')

# Plot the tangent plane
ax.plot_surface(x_tan, y_tan, z_tan, color='r', alpha=0.5, rstride=100, cstride=100, edgecolor='none')

# Highlight the point of tangency
ax.scatter(1, 0, 1, color='k', s=50)

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Tangent Plane to the Paraboloid at Point P(1, 0, 1)')

# Show the plot
plt.show()
```

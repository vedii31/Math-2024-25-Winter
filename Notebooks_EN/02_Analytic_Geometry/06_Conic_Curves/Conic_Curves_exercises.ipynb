# Problem: Find the Equation of a Circle

Find the equation of a circle with:  
- Center: $A(1, 2)$  
- Radius: $r = 3$

---

## Solution

### 1. General Equation of a Circle
The equation of a circle with center $(h, k)$ and radius $r$ is given by:  

$$
(x - h)^2 + (y - k)^2 = r^2
$$

### 2. Substitute the Given Values
Here, the center is $A(1, 2)$, so $h = 1$ and $k = 2$. The radius is $r = 3$. Substituting these values:  

$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

Simplify the equation:  

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---


```python
import matplotlib.pyplot as plt
import numpy as np

# Circle parameters
h, k = 1, 2  # Center of the circle
r = 3  # Radius of the circle

# Generate points for the circle
theta = np.linspace(0, 2 * np.pi, 100)
x = h + r * np.cos(theta)  # x-coordinates
y = k + r * np.sin(theta)  # y-coordinates

# Plot the circle
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r"$(x-1)^2 + (y-2)^2 = 9$", color="blue")
plt.scatter(h, k, color="red", label="Center A(1, 2)")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Add grid, labels, and title
plt.gca().set_aspect('equal', adjustable='box')  # Equal aspect ratio
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Graph of the Circle")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Problem: Find the Equation of a Parabola

The parabola intersects the \( x \)-axis at points \( x = 2 \) and \( x = 4 \), and passes through the point \( (3, 1) \). Find the equation of the parabola.

---

## Solution

### 1. General Equation of a Parabola
The general form of the equation of a parabola is:

$$
y = a(x - p)(x - q)
$$

where:
- \( p \) and \( q \) are the \( x \)-coordinates of the points where the parabola intersects the \( x \)-axis, and
- \( a \) is a constant that will be determined.

### 2. Substitute the Given Points
The parabola intersects the \( x \)-axis at \( x = 2 \) and \( x = 4 \). So, \( p = 2 \) and \( q = 4 \).

Substitute these values into the equation:

$$
y = a(x - 2)(x - 4)
$$

### 3. Use the Given Point to Find \( a \)
The parabola passes through the point \( (3, 1) \), meaning when \( x = 3 \), \( y = 1 \). Substitute \( x = 3 \) and \( y = 1 \) into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplify:

$$
1 = a(1)(-1)
$$

$$
1 = -a
$$

So, \( a = -1 \).

### 4. Write the Final Equation
Substitute \( a = -1 \) into the equation:

$$
y = -1(x - 2)(x - 4)
$$

Simplify:

$$
y = -(x - 2)(x - 4)
$$

---

### Final Answer
The equation of the parabola is:

$$
y = -(x - 2)(x - 4)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Parabola parameters
a = -1  # From the equation y = -(x - 2)(x - 4)
p, q = 2, 4  # Intersects the x-axis at x=2 and x=4

# Generate x values
x = np.linspace(0, 6, 400)

# Parabola equation: y = -1(x - 2)(x - 4)
y = a * (x - p) * (x - q)

# Plot the parabola
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r"$y = -(x - 2)(x - 4)$", color="blue")
plt.scatter([2, 4], [0, 0], color="red", label="Intersects at $(2,0)$ and $(4,0)$")
plt.scatter(3, 1, color="green", label="Point $(3, 1)$")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Add grid, labels, and title
plt.gca().set_aspect('equal', adjustable='box')  # Equal aspect ratio
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Graph of the Parabola")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Problem: Find the Center of the Ellipse

Find the center of the ellipse with the equation:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

---

## Solution

### 1. Rearranging the Equation
We start by rewriting the given equation:

$$
x^2 - 4x + 4y^2 - 16y + 16 = 0
$$

Now, group the \(x\)-terms and \(y\)-terms together:

$$
(x^2 - 4x) + 4(y^2 - 4y) + 16 = 0
$$

### 2. Completing the Square
To complete the square for the \(x\)-terms and \(y\)-terms, we follow these steps:

- For \(x^2 - 4x\), add and subtract \( \left(\frac{-4}{2}\right)^2 = 4 \).
- For \(y^2 - 4y\), factor out the 4 and add and subtract \( \left(\frac{-4}{2}\right)^2 = 4 \) inside the parentheses.

Now, we have:

$$
(x^2 - 4x + 4) + 4(y^2 - 4y + 4) = 16 - 16 + 16
$$

### 3. Simplify the Equation
Simplify the equation:

$$
(x - 2)^2 + 4(y - 2)^2 = 16
$$

### 4. Identify the Center
The equation of the ellipse is now in standard form:

$$
\frac{(x - 2)^2}{16} + \frac{(y - 2)^2}{4} = 1
$$

From this, we can see that the center of the ellipse is at:

$$
(2, 2)
$$

---

### Final Answer
The center of the ellipse is:

$$
(2, 2)
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Ellipse parameters (center: (2, 2), equation: (x - 2)^2/16 + (y - 2)^2/4 = 1)
center_x, center_y = 2, 2
a = 4  # Semi-major axis (sqrt(16))
b = 2  # Semi-minor axis (sqrt(4))

# Parametric equations for the ellipse
theta = np.linspace(0, 2 * np.pi, 400)  # Angle values for the parametric equation
x = center_x + a * np.cos(theta)  # x = center_x + a * cos(theta)
y = center_y + b * np.sin(theta)  # y = center_y + b * sin(theta)

# Plot the ellipse
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r"$\frac{(x - 2)^2}{16} + \frac{(y - 2)^2}{4} = 1$", color="blue")

# Mark the center of the ellipse
plt.scatter(center_x, center_y, color="red", label="Center (2, 2)", zorder=5)

# Set axis properties
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')  # Equal scaling for x and y axes
plt.grid(True)

# Add labels, title, and legend
plt.title("Graph of the Ellipse")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Problem: Find the Slope of the Tangent Line

Find the slope \( m \) of the line \( y = mx - 5 \) that is tangent to the circle with the equation:

$$
x^2 + y^2 = 1
$$

---

## Solution

### 1. Equation of the Circle and Line
The equation of the circle is:

$$
x^2 + y^2 = 1
$$

The equation of the line is:

$$
y = mx - 5
$$

---

### 2. Find the Point of Tangency
For the line to be tangent to the circle, it must intersect the circle at exactly one point. This means that the system of equations formed by the circle and the line must have exactly one solution.

Substitute the equation of the line \( y = mx - 5 \) into the equation of the circle \( x^2 + y^2 = 1 \):

$$
x^2 + (mx - 5)^2 = 1
$$

Simplify:

$$
x^2 + (m^2x^2 - 10mx + 25) = 1
$$

Combine like terms:

$$
(1 + m^2)x^2 - 10mx + 25 = 1
$$

Simplify further:

$$
(1 + m^2)x^2 - 10mx + 24 = 0
$$

---

### 3. Condition for a Single Solution
For the line to be tangent to the circle, this quadratic equation must have exactly one solution. This happens when the discriminant is zero. The discriminant \( \Delta \) for a quadratic equation \( ax^2 + bx + c = 0 \) is given by:

$$
\Delta = b^2 - 4ac
$$

In our case, \( a = 1 + m^2 \), \( b = -10m \), and \( c = 24 \). Set the discriminant to zero:

$$
(-10m)^2 - 4(1 + m^2)(24) = 0
$$

Simplify:

$$
100m^2 - 96(1 + m^2) = 0
$$

Distribute:

$$
100m^2 - 96 - 96m^2 = 0
$$

Combine like terms:

$$
4m^2 = 96
$$

Solve for \( m^2 \):

$$
m^2 = 24
$$

Take the square root of both sides:

$$
m = \sqrt{24} = 2\sqrt{6}
$$

Since \( m > 0 \), we have:

$$
m = 2\sqrt{6}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Circle equation: x^2 + y^2 = 1
# Line equation: y = mx - 5, with m = 2 * sqrt(6)

m = 2 * np.sqrt(6)  # The slope of the tangent line
x_vals = np.linspace(-1.5, 1.5, 400)  # x values for plotting

# Equation of the line: y = mx - 5
y_line = m * x_vals - 5

# Equation of the circle: x^2 + y^2 = 1
# Parametric equations for the circle
theta = np.linspace(0, 2 * np.pi, 400)
x_circle = np.cos(theta)
y_circle = np.sin(theta)

# Plotting
plt.figure(figsize=(6, 6))

# Plot the circle
plt.plot(x_circle, y_circle, label="Circle: $x^2 + y^2 = 1$", color="blue")

# Plot the tangent line
plt.plot(x_vals, y_line, label=f"Tangent Line: $y = {m:.2f}x - 5$", color="red")

# Set the axis to be equal
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

# Labels, title and legend
plt.title("Circle and Tangent Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Problem: Find the Intersection Points of the Hyperbola and the Ellipse

Find the intersection points of the hyperbola with the equation:

$$
x^2 - y^2 = 1
$$

and the ellipse with the equation:

$$
x^2 + 4y^2 = 6
$$

---

## Solution

### 1. Write the Equations
The equations of the hyperbola and ellipse are:

**Hyperbola:**
$$
x^2 - y^2 = 1
$$

**Ellipse:**
$$
x^2 + 4y^2 = 6
$$

### 2. Solve the System of Equations
We need to solve the system of equations:

$$
x^2 - y^2 = 1 \tag{1}
$$
$$
x^2 + 4y^2 = 6 \tag{2}
$$

### 3. Solving for the Intersection Points
By solving the above system of equations, we find the intersection points of the hyperbola and the ellipse.

### 4. Simplify and Find Solutions
The intersection points are:

$$
(-1.41421356237310, -1.00000000000000), \quad (-1.41421356237310, 1.00000000000000)
$$
$$
(1.41421356237310, -1.00000000000000), \quad (1.41421356237310, 1.00000000000000)
$$

These are the points where the hyperbola and ellipse intersect.

---

### Final Answer
The intersection points of the hyperbola and ellipse are:

$$
(-1.41421356237310, -1.00000000000000), (-1.41421356237310, 1.00000000000000)
$$
$$
(1.41421356237310, -1.00000000000000), (1.41421356237310, 1.00000000000000)
$$

---


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve

# Define variables
x, y = symbols('x y')

# Define the equations
eq1 = Eq(x**2 - y**2, 1)  # Hiperbola: x^2 - y^2 = 1
eq2 = Eq(x**2 + 4*y**2, 6)  # Elips: x^2 + 4y^2 = 6

# Solve the system of equations
solutions = solve((eq1, eq2), (x, y))
solutions = [(sol[0].evalf(), sol[1].evalf()) for sol in solutions]

# Display the solutions (intersection points)
print("Intersection points of the hyperbola and ellipse:")
for sol in solutions:
    print(f"({sol[0]}, {sol[1]})")

# For plotting the curves and intersection points
x_vals = np.linspace(-3, 3, 400)

# Avoid invalid sqrt by checking the values
y_vals_hyperbola_pos = np.sqrt(np.maximum(x_vals**2 - 1, 0))  # Ensure the argument is non-negative
y_vals_hyperbola_neg = -np.sqrt(np.maximum(x_vals**2 - 1, 0))
y_vals_ellipse_pos = np.sqrt(np.maximum((6 - x_vals**2) / 4, 0))
y_vals_ellipse_neg = -np.sqrt(np.maximum((6 - x_vals**2) / 4, 0))

# Plotting the hyperbola and ellipse
plt.figure(figsize=(6, 6))

# Plot the hyperbola
plt.plot(x_vals, y_vals_hyperbola_pos, label="Hyperbola: $x^2 - y^2 = 1$", color="blue")
plt.plot(x_vals, y_vals_hyperbola_neg, color="blue")
# Plot the ellipse
plt.plot(x_vals, y_vals_ellipse_pos, label="Ellipse: $x^2 + 4y^2 = 6$", color="red")
plt.plot(x_vals, y_vals_ellipse_neg, color="red")

# Plot the intersection points
for sol in solutions:
    plt.scatter(sol[0], sol[1], color="green", zorder=5)
    plt.scatter(sol[0], -sol[1], color="green", zorder=5)

# Set the axis to be equal
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

# Labels, title, and legend
plt.title("Intersection of Hyperbola and Ellipse")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Problem: Find the Distance Between the Branches of the Hyperbola

For the given hyperbola with the equation:

$$
x^2 - y^2 = 1
$$

We need to find the distance between its branches.

---

## Solution

### 1. Equation of the Hyperbola
The standard form of the hyperbola is:

$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$

The given equation is:

$$
x^2 - y^2 = 1
$$

This is already in standard form with \(a^2 = 1\) and \(b^2 = 1\).

### 2. Distance Between the Branches
For a hyperbola of the form \( \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 \), the distance between its branches along the \(y = 0\) axis (on the x-axis) is given by:

$$
2a
$$

In our case, \(a^2 = 1\), so \(a = 1\).

Thus, the distance between the branches of the hyperbola is:

$$
2a = 2 \times 1 = 2
$$

---

### Final Answer
The distance between the branches of the hyperbola is:

$$
2
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values for plotting
x_vals = np.linspace(-3, 3, 400)

# Hyperbola equation: x^2 - y^2 = 1
# Solve for y: y = ±sqrt(x^2 - 1)
y_vals_pos = np.sqrt(np.maximum(x_vals**2 - 1, 0))  # Positive branch of the hyperbola
y_vals_neg = -np.sqrt(np.maximum(x_vals**2 - 1, 0))  # Negative branch of the hyperbola

# Calculate the distance between the branches at y = 0 (on the x-axis)
distance_between_branches = 2 * 1  # From the equation, a = 1, so distance = 2 * a

# Plotting the hyperbola
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals_pos, label="Positive branch of hyperbola", color="blue")
plt.plot(x_vals, y_vals_neg, label="Negative branch of hyperbola", color="blue")

# Mark the branches' distance at x=1 and x=-1
plt.scatter([-1, 1], [0, 0], color="red", zorder=5, label="Distance points")

# Set the axis to be equal
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)

# Labels, title, and legend
plt.title(f"Distance Between the Branches of the Hyperbola: {distance_between_branches}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Display the calculated distance
print(f"Distance between the branches of the hyperbola: {distance_between_branches}")
```

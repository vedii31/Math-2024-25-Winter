### Taylor Series

### Problem
Compute the Taylor series for $ f(x) = \cos(x) $, centered at $ x = 0 $, up to the 4th degree.


### Taylor Series Formula
The Taylor series is given by:

$$
f(x) = \sum_{n=0}^\infty \frac{f^{(n)}(a)}{n!}(x-a)^n
$$

For this case:
- $f(x) = \cos(x)$
- Center $ a = 0 $
- Expand up to 4th degree $( n \leq 4 \$.


### Step-by-Step Calculation

### 1. Function and Derivatives
Compute the derivatives of $ \cos(x) $:
- $ f(x) = \cos(x) $
- $ f'(x) = -\sin(x) $
- $ f''(x) = -\cos(x) $
- $ f^{(3)}(x) = \sin(x) $
- $ f^{(4)}(x) = \cos(x) $

### 2. Evaluate Derivatives at $ x = 0 $
Substitute $ x = 0 $ into each derivative:
- $ f(0) = \cos(0) = 1 $
- $ f'(0) = -\sin(0) = 0 $
- $ f''(0) = -\cos(0) = -1 $
- $ f^{(3)}(0) = \sin(0) = 0 $
- $ f^{(4)}(0) = \cos(0) = 1 $

### 3. Write the Terms of the Series
Substitute these values into the Taylor series formula:

$$
f(x) \approx f(0) + \frac{f'(0)}{1!}x + \frac{f''(0)}{2!}x^2 + \frac{f^{(3)}(0)}{3!}x^3 + \frac{f^{(4)}(0)}{4!}x^4
$$

Substituting:
- $ f(0) = 1 $
- $ f'(0) = 0 $
- $ f''(0) = -1 $
- $ f^{(3)}(0) = 0 $
- $ f^{(4)}(0) = 1 $

The series becomes:
$$
f(x) \approx 1 + 0 \cdot x + \frac{-1}{2!}x^2 + 0 \cdot x^3 + \frac{1}{4!}x^4
$$

### 4. Simplify the Terms
Simplify each nonzero term:
- $ \frac{-1}{2!} = -\frac{1}{2} $
- $ \frac{1}{4!} = \frac{1}{24} $

The series becomes:
$$
f(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

---

## Final Answer
The Taylor series expansion of $ \cos(x) $ centered at $ x = 0 $ up to the 4th degree is:

$$
f(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

---

## Explanation
- This expansion approximates $ \cos(x) $ near $ x = 0 $.
- Higher-degree terms (e.g., $ x^6, x^8, \ldots $) are omitted since we only expand to the 4th degree.
- The approximation improves near $ x = 0 $ and becomes less accurate as $ x $ moves further away.

```python
import sympy as sp

# Define the variable and function
x = sp.symbols('x')
f = sp.cos(x)

# Find the Taylor series expansion up to the 4th degree
taylor_f = sp.series(f, x, 0, 5)  # Expansion around x = 0, up to degree 4

# Display the result
print("Taylor series for f(x) = cos(x) around x = 0 up to 4th degree:")
print(taylor_f)
```

### Solution for Tangent Line to $ f(x) = e^{\sin(x)} $ at $ x_0 = \pi $


### 1. Tangent Line Formula
The equation of the tangent line at $ x_0 $ is given by:

$$
y = f(x_0) + f'(x_0)(x - x_0)
$$

Where:
- $ f(x_0) $: The value of the function at $ x_0 $,
- $ f'(x_0) $: The slope of the function at $ x_0 $.


### 2. Evaluate $ f(x_0) $
The function is:
$$
f(x) = e^{\sin(x)}
$$

Substitute $ x_0 = \pi $:
$$
f(x_0) = f(\pi) = e^{\sin(\pi)}
$$

Since $ \sin(\pi) = 0 $:
$$
f(\pi) = e^0 = 1
$$

Thus, the value of the function at $ x_0 = \pi $ is:
$$
f(\pi) = 1
$$


### 3. Differentiate $ f(x) $
To compute $ f'(x) $, differentiate $f(x) = e^{\sin(x)} $ using the **chain rule**:
$$
f'(x) = \frac{d}{dx}\left(e^{\sin(x)}\right)
$$

The chain rule states:
$$
\frac{d}{dx}e^u = e^u \cdot \frac{du}{dx}
$$

Here:
- $ u = \sin(x) $,
- $ \frac{du}{dx} = \cos(x) $.

Thus:
$$
f'(x) = e^{\sin(x)} \cdot \cos(x)
$$

---

## 4. Evaluate $ f'(x_0) $
At $ x_0 = \pi $, substitute $ x = \pi $ into $ f'(x) $:
$$
f'(x_0) = f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi)
$$

From trigonometric values:
- $ \sin(\pi) = 0 $,
- $ \cos(\pi) = -1 $.

Thus:
$$
f'(\pi) = e^0 \cdot (-1) = 1 \cdot (-1) = -1
$$

The slope of the tangent line at $ x_0 = \pi $ is:
$$
f'(\pi) = -1
$$


### 5. Write the Tangent Line Equation
Substitute the known values into the tangent line formula:
$$
y = f(x_0) + f'(x_0)(x - x_0)
$$

With:
- $ f(x_0) = 1 $,
- $ f'(x_0) = -1 $,
- $ x_0 = \pi $,

The equation becomes:
$$
y = 1 + (-1)(x - \pi)
$$

Simplify:
$$
y = 1 - x + \pi
$$



### Final Answer
The equation of the tangent line is:
$$
y = -x + \pi + 1
$$


### Explanation of Each Step
1. **Finding $ f(x_0) $:**
   This evaluates the function $ f(x) $ at the specific point $ x_0 = \pi $ to determine the point of tangency.

2. **Differentiating $ f(x) $:**
   This step computes the slope of the function at any $ x $. For $ e^{\sin(x)} $, the chain rule is used to correctly account for the derivative of $ \sin(x) $.

3. **Evaluating $ f'(x_0) $:**
   The slope of the tangent line is computed by substituting $ x_0 = \pi $ into the derivative.

4. **Writing the tangent line:**
   The tangent line formula combines the point of tangency $( x_0, f(x_0) )$ and the slope $( f'(x_0) )$ to construct the equation.

5. **Simplification:**
   The equation is simplified into standard linear form for easier interpretation.

This detailed explanation ensures clarity in each calculation and highlights the step-by-step process.


```python
import sympy as sp

# Define the variable and the function
x = sp.symbols('x')
f = sp.exp(sp.sin(x))

# Point of tangency
x0 = sp.pi

# Compute the derivative of f(x)
f_prime = sp.diff(f, x)

# Evaluate f(x) and f'(x) at x0
f_x0 = f.subs(x, x0)  # f(x0)
f_prime_x0 = f_prime.subs(x, x0)  # f'(x0)

# Define the tangent line equation
tangent_line = f_prime_x0 * (x - x0) + f_x0

# Display the results
print("Function: f(x) = e^(sin(x))")
print("Derivative: f'(x) =", f_prime)
print("f(pi):", f_x0)
print("f'(pi):", f_prime_x0)
print("Tangent line equation: y =", tangent_line)
```

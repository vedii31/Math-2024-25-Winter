# What is Differential Equations?

Differential equations are mathematical equations that relate a function \( y(x) \) to its derivatives. These equations describe how \( y(x) \) changes with respect to \( x \). They are widely used in fields such as physics, engineering, biology, and economics to model real-world phenomena.

#### General Form of a Differential Equation:
$$
F\left(x, y, y', y'', \dots, y^{(n)}\right) = 0
$$

Where:
- \( y = y(x) \) is the unknown function.
- \( y', y'', \dots, y^{(n)} \) are the derivatives of \( y(x) \) with respect to \( x \).
- \( F \) is a function that relates \( x \), \( y \), and its derivatives.

#### Types of Differential Equations:
1. **Ordinary Differential Equations (ODEs):**
   - Contains derivatives of a function \( y(x) \) with respect to one variable \( x \).
   - Example:
     $$
     \frac{dy}{dx} + 2y = 0
     $$

2. **Partial Differential Equations (PDEs):**
   - Contains partial derivatives of a function with respect to multiple variables.
   - Example:
     $$
     \frac{\partial u}{\partial t} = c^2 \frac{\partial^2 u}{\partial x^2}
     $$

#### Order and Degree:
- **Order:** The highest derivative in the equation.
- **Degree:** The power of the highest derivative (after clearing any fractional or root terms).

#### Example of a First-Order ODE:
$$
\frac{dy}{dx} + p(x)y = q(x)
$$
This is called a **linear first-order ODE**, where \( p(x) \) and \( q(x) \) are functions of \( x \).

Differential equations are solved using various methods such as:
- Separation of Variables
- Integrating Factors
- Homogeneous and Non-Homogeneous Solutions
- Laplace Transform

# Examples

1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y$
   - $y'(x) = \frac{1}{2y(x)}$
  
2. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$
   - $\frac{dy}{dx} = \frac{y}{x}$
   - $\frac{dy}{dx} = xy$

3. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

   * $\frac{d^2\,y(x)}{dx^2} = -\omega^2 y(x)$.

4. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

$$

$$


### Final Answer:

## 3/1

We are tasked with solving the equation:
$$
y''(x) + y'(x) = 0
$$
with the boundary conditions \( y(0) = 2 \) and \( y'(0) = -1 \).

#### Step 1: Solve the homogeneous equation
The given equation is a linear homogeneous second-order differential equation. To solve it, first find the characteristic equation:
$$
r^2 + r = 0
$$

Factoring the equation:
$$
r(r + 1) = 0
$$

So, the solutions for \( r \) are:
$$
r_1 = 0, \quad r_2 = -1
$$

Thus, the general solution to the differential equation is:
$$
y(x) = C_1 e^{r_1 x} + C_2 e^{r_2 x} = C_1 + C_2 e^{-x}
$$

#### Step 2: Apply the boundary conditions

1. **Using the first boundary condition \( y(0) = 2 \):**
   $$
   y(0) = C_1 + C_2 e^{0} = C_1 + C_2 = 2
   $$

2. **Using the second boundary condition \( y'(0) = -1 \):**
   First, calculate the derivative of \( y(x) \):
   $$
   y'(x) = -C_2 e^{-x}
   $$

   Now, apply the boundary condition:
   $$
   y'(0) = -C_2 e^{0} = -C_2 = -1
   $$

   So, $ C_2 = 1 $.

#### Step 3: Solve for the constants

From the first boundary condition \( C_1 + C_2 = 2 \), we have:
$$
C_1 + 1 = 2 \quad \Rightarrow \quad C_1 = 1
$$

#### Step 4: Write the final solution
Thus, the solution to the differential equation is:
$$
y(x) = 1 + e^{-x}
$$

### Final Answer:
The solution to the differential equation \( y''(x) + y'(x) = 0 \) with the given boundary conditions is:
$$
y(x) = 1 + e^{-x}
$$


## 3/2

We are tasked with solving the equation:
$$
y''(x) - y(x) = 0
$$
with the boundary conditions \( y(0) = 2 \) and \( y'(0) = 0 \).

#### Step 1: Solve the homogeneous equation
The given equation is a linear homogeneous second-order differential equation. To solve it, first find the characteristic equation:
$$
r^2 - 1 = 0
$$

Factoring the equation:
$$
(r - 1)(r + 1) = 0
$$

So, the solutions for \( r \) are:
$$
r_1 = 1, \quad r_2 = -1
$$

Thus, the general solution to the differential equation is:
$$
y(x) = C_1 e^{r_1 x} + C_2 e^{r_2 x} = C_1 e^x + C_2 e^{-x}
$$

#### Step 2: Apply the boundary conditions

1. **Using the first boundary condition $ y(0) = 2 $:**
   $$
   y(0) = C_1 e^{0} + C_2 e^{0} = C_1 + C_2 = 2
   $$

2. **Using the second boundary condition \( y'(0) = 0 \):**
   First, calculate the derivative of \( y(x) \):
   $$
   y'(x) = C_1 e^x - C_2 e^{-x}
   $$

   Now, apply the boundary condition:
   $$
   y'(0) = C_1 e^{0} - C_2 e^{0} = C_1 - C_2 = 0
   $$

   So, \( C_1 = C_2 \).

#### Step 3: Solve for the constants

From the first boundary condition \( C_1 + C_2 = 2 \) and \( C_1 = C_2 \), we have:
$$
2C_1 = 2 \quad \Rightarrow \quad C_1 = 1
$$

Thus, \( C_2 = 1 \).

#### Step 4: Write the final solution
Therefore, the solution to the differential equation is:
$$
y(x) = e^x + e^{-x}
$$

### Final Answer:
The solution to the differential equation \( y''(x) - y(x) = 0 \) with the given boundary conditions is:
$$
y(x) = e^x + e^{-x}
$$

## 4/1

We are tasked with checking if the function $ \psi(t, x) = A \cos(\omega t + kx) $ is a solution to the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0
$$
where \( v = \frac{\omega}{k} \) is the wave speed.

#### Step 1: Compute the second derivative of $ \psi(t, x) $ with respect to time \( t \)

The given function is:
$$
\psi(t, x) = A \cos(\omega t + kx)
$$

First, compute the first derivative of $ \psi(t, x) $ with respect to \( t \):
$$
\frac{\partial \psi(t, x)}{\partial t} = -A \omega \sin(\omega t + kx)
$$

Now, compute the second derivative with respect to \( t \):
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
$$

#### Step 2: Compute the second derivative of $ \psi(t, x) $ with respect to space \( x \)

Now, compute the first derivative of $ \psi(t, x) $ with respect to \( x \):
$$
\frac{\partial \psi(t, x)}{\partial x} = -A k \sin(\omega t + kx)
$$

Next, compute the second derivative with respect to \( x \):
$$
\frac{\partial^2 \psi(t, x)}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$$

#### Step 3: Substitute into the wave equation

Substitute the second derivatives into the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = -A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx))
$$

Simplify the expression:
$$
-A \omega^2 \cos(\omega t + kx) + A v^2 k^2 \cos(\omega t + kx)
$$

Factor out $ A \cos(\omega t + kx) $:
$$
A \cos(\omega t + kx) \left( v^2 k^2 - \omega^2 \right)
$$

#### Step 4: Use the relationship $ v = \frac{\omega}{k} $

Recall that $ v = \frac{\omega}{k} $, so $ v^2 = \frac{\omega^2}{k^2} $. Substitute this into the equation:
$$
A \cos(\omega t + kx) \left( \frac{\omega^2}{k^2} k^2 - \omega^2 \right)
$$

Simplify:
$$
A \cos(\omega t + kx) \left( \omega^2 - \omega^2 \right) = 0
$$

Thus, the equation is satisfied:
$$
0 = 0
$$

### Final Conclusion:
The function $ \psi(t, x) = A \cos(\omega t + kx) $ **is a solution** to the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0
$$

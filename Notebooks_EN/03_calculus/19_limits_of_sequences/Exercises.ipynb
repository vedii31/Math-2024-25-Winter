## Limits of Sequences

1. Calculate:
   - $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$

   - $\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}$

2. Prove using the squeeze theorem:
   - $\displaystyle\lim_{n \to \infty} \frac{\sin(n)}{n}$

4. Find the limit of the sequence:
   - $a_n = (1+\frac{1}{n})^n$

## 1. Calculate :

We are asked to calculate the following limits:

### 1. First Limit:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

#### Solution:

We can simplify the expression by dividing both the numerator and the denominator by $n^2$, which is the highest degree term:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \lim_{n \to \infty} \frac{n^2(1 + \frac{3}{n})}{n^2(2 - \frac{2}{n})}
$$

Simplifying by canceling $n^2$:

$$
= \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
$$

As $n \to \infty$, both $\frac{3}{n}$ and $\frac{2}{n}$ approach zero, so we get:

$$
= \frac{1 + 0}{2 - 0} = \frac{1}{2}
$$

Thus, the first limit is:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}
$$

---

### 2. Second Limit:

$$
\lim_{n \to \infty} \frac{2n + 3}{3n^3 - 1}
$$

#### Solution:

We simplify the expression by dividing both the numerator and the denominator by $n^3$, which is the highest degree term:

$$
\lim_{n \to \infty} \frac{2n + 3}{3n^3 - 1} = \lim_{n \to \infty} \frac{n(2 + \frac{3}{n})}{n^3(3 - \frac{1}{n^3})}
$$

Simplifying by canceling $n^3$:

$$
= \lim_{n \to \infty} \frac{2 + \frac{3}{n}}{3n^2(3 - \frac{1}{n^3})}
$$

As $n \to \infty$, both $\frac{3}{n}$ and $\frac{1}{n^3}$ approach zero, so we get:

$$
= \frac{2 + 0}{3 \cdot \infty} = 0
$$

Thus, the second limit is:

$$
\lim_{n \to \infty} \frac{2n + 3}{3n^3 - 1} = 0
$$

---

### Final Answers:

1. $$\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}$$
2. $$\lim_{n \to \infty} \frac{2n + 3}{3n^3 - 1} = 0$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions for the sequences
def limit1(n):
    return (n**2 + 3*n) / (2*n**2 - 2*n)

def limit2(n):
    return (2*n + 3) / (3*n**3 - 1)

# Values of n
n_values = np.arange(1, 1000)
limit1_values = limit1(n_values)
limit2_values = limit2(n_values)

# Plotting the sequences to observe their behavior as n increases
plt.plot(n_values, limit1_values, label=r"$\frac{n^2 + 3n}{2n^2 - 2n}$", color='b')
plt.axhline(y=1/2, color='r', linestyle='--', label="Limit = 1/2")

plt.plot(n_values, limit2_values, label=r"$\frac{2n + 3}{3n^3 - 1}$", color='g')
plt.axhline(y=0, color='r', linestyle='--', label="Limit = 0")

plt.xlabel('n')
plt.ylabel('Limit Value')
plt.title('Convergence of the Sequences to their Limits')
plt.legend()
plt.grid(True)
plt.show()

# Displaying the values of the sequences for large n
limit1_value = limit1(10000)  # Approximation for large n
limit2_value = limit2(10000)  # Approximation for large n

print(f"Approximation of the first limit for n=10000: {limit1_value}")
print(f"Approximation of the second limit for n=10000: {limit2_value}")
```

## 2. Squeeze Theorem: Prove the Limit


We are tasked with proving the following limit using the squeeze theorem:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

### Solution:

We begin by applying the Squeeze Theorem. The Squeeze Theorem states that if:

$$
a_n \leq b_n \leq c_n
$$

for all \(n\) in some interval, and

$$
\lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L
$$

then

$$
\lim_{n \to \infty} b_n = L
$$

Now, let's examine the function:

$$
b_n = \frac{\sin(n)}{n}
$$

We know that the sine function is always bounded:

$$
-1 \leq \sin(n) \leq 1
$$

Dividing through by \(n\), where \(n > 0\), we get:

$$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

Now, we take the limit of the bounding functions as \(n \to \infty\):

$$
\lim_{n \to \infty} \frac{-1}{n} = 0
$$
$$
\lim_{n \to \infty} \frac{1}{n} = 0
$$

Since both bounds approach 0 as \(n \to \infty\), by the Squeeze Theorem, we conclude:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

### Final Answer:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence a_n = sin(n) / n
def a_n(n):
    return np.sin(n) / n

# Values of n
n_values = np.arange(1, 1000)
a_n_values = a_n(n_values)

# Plotting the sequence to observe its behavior as n increases
plt.plot(n_values, a_n_values, label=r"$\frac{\sin(n)}{n}$", color='b')
plt.axhline(y=0, color='r', linestyle='--', label="y=0 (Limit)")

plt.xlabel('n')
plt.ylabel(r'$\frac{\sin(n)}{n}$')
plt.title(r'Convergence of the sequence $\frac{\sin(n)}{n}$ to 0')
plt.legend()
plt.grid(True)
plt.show()

# Displaying the value of the sequence for large n
limit_value = a_n(10000)  # Approximation for large n
print(f"Approximation of the limit for n=10000: {limit_value}")
```

# 3. Limit of the Sequence

We are asked to find the limit of the following sequence:

$$
a_n = \left( 1 + \frac{1}{n} \right)^n
$$

### Solution:

This is a classic sequence, and it converges to \( e \), the base of the natural logarithm. We can show this using the definition of \( e \).

We know that:

$$
e = \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n
$$

So, for this sequence:

$$
\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e
$$

### Final Answer:

$$
\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence a_n = (1 + 1/n)^n
def a_n(n):
    return (1 + 1/n)**n

# Values of n
n_values = np.arange(1, 1000)
a_n_values = a_n(n_values)

# Plotting the sequence to observe its behavior as n increases
plt.plot(n_values, a_n_values, label=r"$\left(1 + \frac{1}{n}\right)^n$")
plt.axhline(y=np.e, color='r', linestyle='--', label="e (Limit)")

plt.xlabel('n')
plt.ylabel(r'$a_n$')
plt.title('Convergence of the sequence $(1 + 1/n)^n$ to e')
plt.legend()
plt.grid(True)
plt.show()

# Displaying the value of the sequence for large n
limit_value = a_n(10000)  # Approximation for large n
print(f"Approximation of the limit for n=10000: {limit_value}")
```
